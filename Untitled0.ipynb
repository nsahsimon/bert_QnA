```python
from transformers import pipeline

# retrieving default question answering model
question_answerer = pipeline("question-answering") 

context = "Tom Brady is the greastest quarterback of all time and he plays for the tamper bay in the united states. He has 7 super balls."

result = question_answerer(question="who is the greatest quaterback", context=context)

print(f"answer: {result['answer']}")
```
