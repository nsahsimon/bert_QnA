```python
from transformers import pipeline

# retrieving default question answering model
question_answerer = pipeline("question-answering") 

context = "Tom Brady is the greastest quarterback of all time and he plays for the tamper bay in the united states. He has 7 super balls."
question = "In which team does Tom Brady play?"
result = question_answerer(question=question, context=context)

print(f"answer: {result['answer']}")
```

**Using Tensorflow**

```python
from transformers import AutoTokenizer, TFAutoModelForQuestionAnswering
import tensorflow as tf

model_name = "bert-large-uncased-whole-word-masking-finetuned-squad"

# set the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = TFAutoModelForQuestionAnswering.from_pretrained(model_name)


context = "Tom Brady is the greastest quarterback of all time and he plays for the tamper bay in the united states. He has 7 super balls."
questions = [
    "Who is the best player in the world ",
    "In which team does the best player in the world play?"
    ]

# loop through the questions
for question in questions:
  inputs = tokenizer(question, context, add_special_tokens=True, return_tensors="tf")
  input_ids = inputs["input_ids"].numpy()[0]

  outputs = model(inputs)
  answer_start_scores = outputs.start_logits
  answer_end_scores = outputs.end_logits

  answer_start = tf.argmax(
      answer_start_scores, axis = 1
  ).numpy()[0]

  answer_end = (
      tf.argmax(answer_end_scores, axis=1) + 1
  ).numpy()[0]

  answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(input_ids[answer_start:answer_end]))
  
  print(f"Question: {question}")
  print(f"Answer: {answer}")
```
